- in that case, multivariate analysis helps us to reduce dimension

```python
#import
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df_origin = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NSCLC clinic project/clinical_information.csv')
df = df_origin.copy()
df.describe()
```

```python
df.columns
```

```python
df_X = df[['primary_tumor_pathologic_spread',
           'residual_tumor',
           'sodium',
           'BUN',
           'Creatinine',
           'PLR',
           'pathology-lymphovascularinvasion#-',
           'pathology-visceralpleural#-',]]

df_X2 = df.drop(columns=['2yRFS', 'data_num', 'tumor_stage', 'lymphnode_pathologic_spread'])

df_y = df['2yRFS']#RFS 재발없는 생존율
```

```python
X1_with_const = sm.add_constant(df_X)
model_X1 = sm.Logit(df_y, df_X).fit()
```

```python
print(model_X1.summary())
```

```python
params = model_X1.params
conf = model_X1.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
print(np.exp(conf))
```
